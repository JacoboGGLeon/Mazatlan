```python
import pandas as pd
```

```python
file = '../dataframes/tadolescente_enadis_2017.csv'
```

```python
df = pd.read_csv(filepath_or_buffer=file)
```

```python
df.head()
```

```python
df_copy = df.copy()
```

```python
df_copy.columns
```

```python
df_copy.drop(labels=['Unnamed: 0', 'upm', 'viv_sel', 'hogar', 'n_ren', 'factor_per', 'upm_dis', 'est_dis'], axis=1, inplace=True)
```

```python
df_copy = df_copy.astype('int64')
```

```python
for columns in df_copy.columns:
    print(df_copy[columns].value_counts())
```
